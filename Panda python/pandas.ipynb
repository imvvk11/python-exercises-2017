```python
#whiskies classification proje ct


import numpy as np
import pandas as pd

whisky = pd.read_csv("whiskies.txt")                 #read csv files from txt
whisky["Region"] = pd.read_csv("regions.txt") 
```

```python
whisky.head()   #display first five rows and whiskey.tail() displays last 5 rows
```

```python
whisky.iloc[0:10]
```

```python
whisky.iloc[0:5, 1:5]  #corresponds to rows, columns
```

```python
whisky.columns      #display all columns name
```

```python
flavor = whisky.iloc[:, 2:14] #display all rows with specified  columns for selecting flavors in whisky 

flavor
```

```python
#Compute correlation across the columns of the data frame

corr_flavors = pd.DataFrame.corr(flavor)

corr_flavors
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.pcolor(corr_flavors)
plt.colorbar()
plt.savefig("corr_flavors.pdf")
```

```python
plt.show()
```

```python
corr_whisky = pd.DataFrame.corr(flavor.transpose())  #transpose: rows to columns and columns to rows
print(corr_whisky)

plt.figure(figsize=(10,10))
plt.pcolor(corr_whisky)
plt.axis("tight")
plt.colorbar()
plt.savefig("corr_whisky.pdf")
```

```python
plt.show()
```

```python
#clustering whiskies by flavor profile

from sklearn.cluster.bicluster import SpectralCoclustering

model = SpectralCoclustering(n_clusters=6, random_state=0)
model.fit(corr_whisky)
```

```python
model.rows_
```

```python
np.sum(model.rows_, axis=1)    #axis=1 is column
```

```python
np.sum(model.rows_, axis=0)  
```

```python
model.row_labels_
```

```python
#comparing corelation matrices

whisky["Group"] = pd.Series(model.row_labels_,index=whisky.index)

whisky = whisky.ix[np.argsort(model.row_labels_)]
whisky = whisky.reset_index(drop=True)

correlations = pd.DataFrame.corr(whisky.iloc[:,2:14].transpose())

correlations = np.array(correlations)
```

```python
plt.figure(figsize=(14,7))
plt.subplot(121)
plt.pcolor(corr_whisky)
plt.title("original")
plt.axis("tight")
plt.subplot(122)
plt.pcolor(correlations)
plt.title("rearranged")
plt.axis("tight")
plt.savefig("correlations.pdf")
```
